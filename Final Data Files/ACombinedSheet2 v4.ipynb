```python
import requests,pandas as pd
import numpy as np
import json
from pandas import json_normalize

# Display full number versus scientific notation
pd.set_option('display.float_format', lambda x: '%.3f' % x)

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('JMACData02.xlsx', engine='xlsxwriter')
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetAllMakes?format=json';
r = requests.get(url);
print("This provides a list of all the Makes available in vPIC Dataset.")
print()
dict = json.loads(r.text)
Makes = pd.json_normalize(dict['Results']) 
Makes.to_excel(writer, sheet_name='Sheet1-Makes')
Makes
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetMakesForVehicleType/car?format=json';
r = requests.get(url);
print("This returns all the Makes in the vPIC dataset for a specified vehicle type whose name contains CAR.")
print()
dict = json.loads(r.text)
Car_Type_Vehicle = pd.json_normalize(dict['Results']) 
Car_Type_Vehicle.to_excel(writer, sheet_name='Sheet2A-Car_Type_Vehicle')
Car_Type_Vehicle
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetMakesForVehicleType/truck?format=json';
r = requests.get(url);
print("This returns all the Makes in the vPIC dataset for a specified vehicle type whose name contains TRUCK.")
print()
dict = json.loads(r.text)
Truck_Type_Vehicle = pd.json_normalize(dict['Results']) 
Truck_Type_Vehicle.to_excel(writer, sheet_name='Sheet2B-Truck_Type_Vehicle')
Truck_Type_Vehicle
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetMakesForVehicleType/Moto?format=json';
r = requests.get(url);
print("This returns all the Makes in the vPIC dataset for a specified vehicle type whose name contains MOTO.")
print()
dict = json.loads(r.text)
Moto_Vehicle = pd.json_normalize(dict['Results']) 
Moto_Vehicle.to_excel(writer, sheet_name='Sheet2C-Moto_Vehicle')
Moto_Vehicle
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetMakesForVehicleType/Trailer?format=json';
r = requests.get(url);
print("This returns all the Makes in the vPIC dataset for a specified vehicle type whose name contains TRAILER.")
print()
dict = json.loads(r.text)
Trailer_Vehicle = pd.json_normalize(dict['Results']) 
Trailer_Vehicle.to_excel(writer, sheet_name='Sheet2D-Trailer_Vehicle')
Trailer_Vehicle
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetMakesForVehicleType/Vehicle?format=json';
r = requests.get(url);
print("This returns all the Makes in the vPIC dataset for a specified vehicle type whose name contains VEHICLE.")
print()
dict = json.loads(r.text)
VEH_Vehicle = pd.json_normalize(dict['Results']) 
VEH_Vehicle.to_excel(writer, sheet_name='Sheet2E-VEH_Vehicle')
VEH_Vehicle
```

```python
VTypes = pd.DataFrame()
VTypes = pd.concat([Car_Type_Vehicle, Truck_Type_Vehicle, Moto_Vehicle, Trailer_Vehicle, VEH_Vehicle], axis=0)
VTypes = VTypes.sort_values('MakeId')
VTypes.index = np.arange(1,len(VTypes)+1)
VTypes.to_excel(writer, sheet_name='Sheet2Z-VehicleTypes')
VTypes
```

```python
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetEquipmentPlantCodes?year=2015&equipmentType=1&reportType=all&format=json';
r = requests.get(url);
print("This provides a list of all the Equipment Plants and Codes in vPIC Dataset.")
print()
dict = json.loads(r.text)
Eqip_Plant = pd.json_normalize(dict['Results']) 
Eqip_Plant = Eqip_Plant.replace(to_replace = "UNITED STATES (USA)", value="UNITED STATES")
Eqip_Plant = Eqip_Plant.replace(to_replace = "United States (USA)", value="UNITED STATES")
Eqip_Plant = Eqip_Plant.replace(to_replace = "UNITED KINGDOM (UK)", value="UNITED KINGDOM")
Eqip_Plant = Eqip_Plant.replace(to_replace = "MYANMAR (BURMA)", value="MYANMAR")
Eqip_Plant['Country'] = Eqip_Plant['Country'].str.upper()
Eqip_Plant['StateProvince'] = Eqip_Plant['StateProvince'].str.upper()
Eqip_Plant = Eqip_Plant.set_index('DOTCode')
Eqip_Plant.to_excel(writer, sheet_name='Sheet4A-Eqip_Plant')
Eqip_Plant
```

```python
EPStatus = pd.DataFrame()
EPStatus = Eqip_Plant.groupby('Status').size()
Eqip_Plant.to_excel(writer, sheet_name='Sheet4B-EP-Status')
EPStatus
```

```python
EPLocation = pd.DataFrame()
EPLocation = Eqip_Plant.groupby('Country').size()

EPLocation.to_excel(writer, sheet_name='Sheet4C-EPLocation')
EPLocation
```

```python
EPAct = pd.DataFrame()
EPActive = pd.DataFrame()

EPAct = Eqip_Plant[Eqip_Plant['Status'] != 'Closed']
EPActive = EPAct.groupby('Country').size()
EPLocation.to_excel(writer, sheet_name='Sheet4D-EP-Active')
EPActive
```
